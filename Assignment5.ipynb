# Learn about magic commands and what they are used for
Magic commands are special commands that enhance the functionality of the Python programming environment and are not standard python syntax.
Magic commands are divided into two:
1.Line magics- start with % and operate on a single line of output
2.Cell magics- start with %% and operate on multiple line of input(the entire cell)
## Differences between a bar chart, countplot and histogram
1.Bar chart
A bar chart uses rectangular bars to represent data and the height of each bar corresponds to the value of a category.
It is used to compare quantities across different categories
The X-Axis represents categories
The Y-Axis represents the value or count for each category
The bars are separated from each other showing the data is discrete and not continuous.
2.Countplot
A countplot is a special type of bar chart that shows the count of observations in each category
The X-Axis represents categories eg 'Male', 'Female'
The Y-Axis represents the count of occurrences for each category
The bars are separated
They are used for visualizing the frequency of categorical variables in a dataset eg, gender distribution
3.Histogram
A histogram is used to show the distribution of numerical variable
It divides the data into bins and counts the number of observations in each bin
The X-Axis represents ranges(bins) of numerical values
The Y-Acis represents the frequency of data points within each range
The bars are adjacent to each other showing that the data is continuous
### Do a blog on data protection, privacy and ethics

The Importance of Data Protection
Data protection refers to the measures and practices implemented to safeguard sensitive information from unauthorized access, misuse, or breaches. With the exponential growth of data collection, organizations now hold vast amounts of personal and sensitive information, ranging from financial records to health data. Protecting this data is not just a legal obligation but also a moral responsibility.
A failure to protect data can lead to privacy breaches, eroding trust between individuals and organizations. Trust is a cornerstone of any relationship, and when it is compromised, the consequences can be severe, including reputational damage, financial losses, and legal repercussions.Robust data protection measures, such as encryption, access controls, and regular audits, are essential to ensure that sensitive information remains secure.
Understanding Data Privacy
Data privacy focuses on the rights of individuals to control how their personal information is collected, used, and shared. It is closely tied to the concept of autonomy, as individuals should have the ability to decide what happens to their data. Ethical data practices ensure that individuals have control over their information and are protected from potential misuse or abuse. One of the key principles of data privacy is consent. Organizations must obtain explicit consent from individuals before collecting or using their data. Transparency is equally importantâ€”users should be informed about how their data will be used and for what purposes. For example, ethical data collection involves ensuring anonymity where necessary and being upfront about data usage.
The Role of Ethics in Data Usage
Ethics in data usage goes beyond compliance with laws and regulations. It involves making decisions that respect individual rights, promote fairness, and avoid harm. Ethical data practices are essential in fostering trust and ensuring that technological advancements do not come at the expense of human dignity.

```python
import pandas as pd
import matplotlib.pyplot as plt

fruits = ['Apples', 'Bananas', 'Cherries', 'Dates']
sales = [400, 350, 300, 450]

plt.figure(figsize=(6,4))
plt.barh(fruits, sales, color='skyblue')
plt.title('Fruit Sales', loc='left')
plt.xlabel('Units Sold')
plt.ylabel('Fruits')
plt.tight_layout
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

scores = [55, 62, 67, 70, 72, 75, 78, 80, 82, 85, 88, 90, 92, 95, 98, 100, 100, 65, 77, 89]
plt.figure(figsize=(8,6))
# Create a histogram
plt.hist(scores, bins=8, color='skyblue', edgecolor='black')

# Add titles and labels
plt.title('Distribution of Math Test Scores')
plt.xlabel('Score Ranges')
plt.ylabel('Number of Students')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample dataset: Heights of students in three classes
class_A = [150, 160, 165, 170, 175, 180, 155, 165, 160, 170]
class_B = [140, 145, 150, 155, 160, 165, 170, 175, 180, 185]
class_C = [155, 160, 165, 170, 175, 180, 185, 190, 195, 200]

# Combine the data into a list
data = [class_A, class_B, class_C]

# Create a boxplot
plt.boxplot(data, labels=['Class A', 'Class B', 'Class C'], patch_artist=True)

# Add titles and labels
plt.title('Height Distribution of Students in Different Classes')
plt.ylabel('Height (cm)')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample dataset: Scores of students in three subjects
data = {
    "Math": [85, 90, 78, 92, 88, 76, 95, 89, 84, 91],
    "Science": [88, 85, 82, 95, 90, 80, 96, 87, 83, 92],
    "English": [78, 82, 85, 88, 84, 79, 91, 86, 80, 87]
}

# Convert the dataset into a NumPy array
scores = np.array([data["Math"], data["Science"], data["English"]])

# Calculate the correlation matrix
correlation_matrix = np.corrcoef(scores)
plt.figure(figsize=(9,8))
# Create a heatmap
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='nearest')

# Add color bar
plt.colorbar(label='Correlation Coefficient')

# Add labels
plt.xticks(ticks=[0, 1, 2], labels=["Math", "Science", "English"])
plt.yticks(ticks=[0, 1, 2], labels=["Math", "Science", "English"])
plt.title("Correlation Matrix Heatmap")

# Show the plot
plt.show()
```
