# 1.Understand the difference between loading (json) and reading (CSV )
JSON stands for JavaScript Object Notation which is structured as a collection of key-value pairs, similar to a dictionary, while a CSV has a tabular format with sets of rows and columns
To load data from  JSON file, you first open the file using python's built in open funtion. Then you pass the file object to the json.load function, which returns a python object representing the contents of the file
Reading CSV involves first importing pandas, then reading it row by row or reading it in a tabular format like a dataframe
## 2.The difference between writing (CSV) and dumping (json)
Writing to a CSV file involves creating a file then writing data in a tabular format with rows and columns, with each field separated by a comma
Dumping JSON involves converting a native data structure into a JSON formatted string and then writing that string to a file
CSV (to_csv()) is best for structured, tabular data.
JSON (json.dump()) is better for nested, hierarchical data.
CSV is easier for spreadsheets and databases, while JSON is common for APIs and configurations.
